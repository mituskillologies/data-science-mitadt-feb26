```python
def add3(x: int = 5, y: int = 10, z: int = 40) -> int:

    """
    this does the addition of 3 numbers

    args:
    x: 1st number
    y: 2nd number
    z: 3rd number

    returns:
    res: addition of x, y, z
    """
    
    res = x + y + z
    return res
```

```python
help(add3)
```
