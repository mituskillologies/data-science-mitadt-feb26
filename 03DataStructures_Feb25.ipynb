```python
# from this list using list comprehension create a new list having names 
# starting with 'R' or 'r'
names = ["riya", "Rohit", "Ankit", 'ansh', "Kunal", "Riddhi"]
```

```python
e = []
for i in names:
    if i.lower().startswith('r'):
        e.append(i)
```

```python
nn = [i for i in names if i.lower().startswith('r')]
nn
```
