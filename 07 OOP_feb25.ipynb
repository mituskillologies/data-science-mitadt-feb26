```python
class Person:

    def __init__(self, name, age, address):  # constructor
        self.name = name
        self.age = age
        self.address = address

    def info(self):
        print(f"name: {self.name}, age: {self.age}, lives in: {self.address}")
```

```python
p1 = Person("Ankit", 29, "Pune")
```

```python
p1.name
```

```python
p1.info()
```
