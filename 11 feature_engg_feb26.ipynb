```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
df.describe()
```

```python
x = df[["Age", "EstimatedSalary"]]
```

```python
x
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel("age")
plt.ylabel("salary")
plt.scatter(df['Age'], df["EstimatedSalary"], c = df["Purchased"])
plt.grid();
```

# normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# create an object
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
# check min and max values from this
```

```python
x_scaled[:, 0].min(), x_scaled[:, 0].max()
```

```python
x_scaled[:, 1].min(), x_scaled[:, 1].max()
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel("age")
plt.ylabel("salary")
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c = df["Purchased"])
plt.grid();
```
