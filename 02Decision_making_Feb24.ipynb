```python
# use nested if only (no 'and' or 'or'/ elif)
```

```python
num = int(input("enter the number"))

if num % 4 == 0:
    if num % 100 == 0:
        if num % 400 == 0:
            print("leap century year")
        else:
            print("century year but not leap ")
    else:
        print("leap year")
else:
    print("not leap")
```

```python
# while
```

```python
count = 1

while count < 11:
    print("pune")
    count += 1
```

```python
# find the addition of numbers -> 1 to 10
```

```python
count = 1
add = 0

while count < 11:
    add = add + count
    count += 1

print("addition is:", add)
```
